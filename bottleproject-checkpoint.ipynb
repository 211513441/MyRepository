```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv.imread('full.jpg')
imgtest=img.copy()

blurred=cv.pyrMeanShiftFiltering(img,31,91)
gray = cv.cvtColor(blurred,cv.COLOR_BGR2GRAY) 
edges = cv.Canny(gray,50,250)
gray=edges
plt.axis("off")
plt.imshow(gray)
plt.show()
ret, thresh = cv.threshold(gray,150,255,cv.THRESH_BINARY_INV+cv.THRESH_OTSU)
```

```python
kernelClose=np.ones((20,20))
maskclose=cv.morphologyEx(thresh,cv.MORPH_OPEN,kernelClose)
```

```python
kernel = np.ones((5,5),np.uint8)
sure_bg = cv.dilate(maskclose,kernel,iterations=1)
# Finding sure foreground area
dist_transform = cv.distanceTransform(maskclose,cv.DIST_L2,5)
ret, sure_fg = cv.threshold(dist_transform,0.7*dist_transform.max(),255,0)

# Finding unknown region
sure_fg = np.uint8(sure_fg)
unknown = cv.subtract(sure_bg,sure_fg)
```

```python
ret, markers = cv.connectedComponents(sure_fg)
# Add one to all labels so that sure background is not 0, but 1
markers = markers+1
# Now, mark the region of unknown with zero
markers[unknown==255] = 0
```

```python
markers = cv.watershed(img,markers)

print(markers)
print(ret)
for i in range(0, ret+1):
   # color = np.random.randint(255,size=3)
    img[markers == i]=[0, 0 ,0]

img[markers==-1]=[255,0,0]
x=img.copy()
```

```python

plt.axis("off")
plt.imshow(x)
plt.show()
forprint2=x.copy()
cv.imwrite('segment.png', forprint2)
x[markers == -1]=[0,0,0]

x[markers == 1 ]=[255,255,255]
forprint4=x.copy()
cv.imwrite('mask.png', forprint4)

plt.axis("off")
plt.imshow(x)
plt.show()
```

```python
img1 = imgtest.copy()
img2=x.copy()
img2[np.where(x == 255)] = img1[np.where(x == 255)]
forprint = cv.cvtColor(img2, cv.COLOR_BGR2RGBA)

plt.axis("off")
plt.imshow(forprint)
plt.show()
```

```python
image = cv.cvtColor(img2, cv.COLOR_BGR2RGBA)
image[np.all(image == [0, 0, 0, 255], axis=2)] = [0, 0, 0, 0]
plt.axis("off")
plt.imshow(image)
plt.show()
forprint3=cv.cvtColor(image, cv.COLOR_BGR2RGBA)
cv.imwrite('purebottle.png', forprint3)
```

```python
glass_MIN = np.array([180, 180, 180,100], np.uint8)    
glass_MAX = np.array([255, 255, 255,255], np.uint8)  
dst = cv.inRange(image, glass_MIN, glass_MAX)
no_glass = cv.countNonZero(dst)
print('The number of Glass pixels is: ' + str(no_glass))
```

```python
imagee = cv.cvtColor(imgtest, cv.COLOR_BGR2RGBA)
all_MIN = np.array([0, 0, 0, 0], np.uint8)    
all_MAX = np.array([255, 255, 255, 255] , np.uint8)  
dst = cv.inRange(imagee, all_MIN, all_MAX)
no_all = cv.countNonZero(dst)
print('The number of all pixels is: ' + str(no_all))
```

```python
z=no_glass/no_all
print(z)
if(z<0.05):
    print('full')
    
if(z>0.1):
    
    print('empty')
    
if(z>0.05 and z<0.1):
    print('half full')
```
